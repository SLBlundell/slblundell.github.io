```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('scraping/player_per_game_salary_2023.csv')
data = data.drop(columns=['Unnamed: 0', '2023/24', '2024/25', '2025/26', '2026/27', '2027/28'])

data.head()
```

```python
corr = data.corr().abs()
corr = corr.loc[corr['2022/23']>.25]
corr.index
variables = list(corr.index)
```

```python
plt.figure(figsize=(20,10))
plt.title("Correlation between Filtered Variables")
sns.heatmap(data=data[variables].corr(), annot=True)
```
