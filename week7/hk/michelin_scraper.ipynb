```python
#import required libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#request content of Wikipedia
michelin_wiki = requests.get('https://en.wikipedia.org/wiki/List_of_Michelin_starred_restaurants_in_Hong_Kong_and_Macau')
print(michelin_wiki)
```

```python
#create BeautifulSoup object
soup = BeautifulSoup(michelin_wiki.content, 'html.parser')
print(soup.prettify())
```

```python
#find correct table
tables = soup.findAll("table", {"class": "wikitable sortable"})
```

```python
#create DataFrame
df = pd.read_html(str(tables))
df = pd.DataFrame(df[0])

df.head()
```

```python
#clean data
df = df.drop(df[df.Territory == "Macau"].index)
df = df.drop("Territory", axis=1)
df.loc[0, "Name"] = df.loc[0, "Name"][43:]
df.columns = df.columns.str.replace('[86345()]', '')
df.columns = df.columns.str.replace('\[\]', '')
```

```python
df.head()
print(df)
```

```python
#find correct table
tables_2 = soup.findAll("table", {"class": "wikitable sortable"})
print(tables_2)
```

```python
df_2 = pd.DataFrame(columns=["name", "territory", "location", "2009", "2010","2011","2012","2013","2014","2015","2016","2017","2018","2019","2020"])
#collect data
for row in tables_2.tbody.find_all('tr'):
    columns = row.find_all('td')

    if(columns != []):
        name = columns[0]
        territory = columns[1]
        location = columns[2]
        twenty_nine = columns[3]
        twenty_ten = columns[4]
        twenty_eleven = columns[5]
        twenty_twelve = columns[6]
        twenty_thirteen = columns[7]
        twenty_foreteen = columns[8]
        twenty_fifteen = columns[9]
        twenty_sixteen = columns[10]
        twenty_seventeen = columns[11]
        twenty_eighteen = columns[12]
        twenty_nineteen = columns[13]
        twenty_twenty = columns[14]

        df_2 = df_2.append({"name": name, "territory": territory, "location": location, "2009": twenty_nine, "2010": twenty_ten, "2011": twenty_eleven,"2012": twenty_twelve, "2013": twenty_thirteen, "2014": twenty_foreteen, "2015": twenty_fifteen,"2016": twenty_sixteen, "2017": twenty_seventeen, "2018": twenty_eighteen, "2019": twenty_nineteen, "2020": twenty_twenty}, ignore_index=True)

df_2.head()
```
