```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

```

```python
salaries = 'https://www.spotrac.com/nba/cap/'

def page(url):
    return requests.get(url)

page(salaries) 
```

```python
soup = BeautifulSoup(page(salaries).text, 'lxml')
soup
```

```python
table = soup.find("table", {"class": "datatable rtable captracker tablesorter tablesorter-default tablesorterfbcec8fdc36758"})
print(table) 
```

```python
headers = []
for i in table.find_all('td'):
    title = i.text
    headers.append(title)

#del headers[7:]
#del headers[0]
print(headers)
```

```python
mydata = pd.DataFrame(columns = headers)

for j in table.find_all("div", "class": "rank"
 row_data = j.find_all(‘td’)
 row = [i.text for i in row_data]
 length = len(mydata)
 mydata.loc[length] = row
```
